```python
#encoded message from book exercise number 3
#code will represent an affine decryption
#@Author = Skyler Moomey
#@Date = 10/18/2020

#APHUO EGEHP PEXOV FKEUH CKVUE CHKVE APHUO
#EGEHU EXOVL EXDKT VGEFT EHFKE UHCKF TZEXO
#VEZDT TVKUE XOVKV ENOHK ZFTEH TEHKQ LEROF
#PVEHP PEXOV ERYKP GERYT GVKEG XDRTE RGAGA

#remove spaces to reduce ambiguity
message = "APHUOEGEHPPEXOVFKEUHCKVUECHKVEAPHUOEGEHUEXOVLEXDKTVGEFTEHFKEUHCKFTZEXOVEZDTTVKUEXOVKVENOHKZFTEHTEHKQLEROFPVEHPPEXOVERYKPGERYTGVKEGXDRTERGAGA"

# v -> E
# O -> H
# X -> T

#function spits out every affine transform of a given message
def affinedecrypt(encrypted):
    
    
    #for affine transform, we need an A and a B, A contained in U(26), B contained in Z_26. How to find A?
    
    for A in range(26):
        
        #determine if A has a multiplicative inverse
        if gcd(A, 26) == 1:
            
            #find A^-1
            A_inv = 0
            for inverse in range(26):
                if(A*inverse)%26 == 1:
                    A_inv = inverse
            
            #B can vary freely from 0 to 25
            for B in range(26):
                
                
                #convert each letter in message to ASCII values, i.e. integers in [65:90]
                digits = [ord(letter) for letter in encrypted]
                
                #subtract 65 to get number in range [0:25]
                digits = [(x-65) for x in digits]
                
                #perform affine transform
                digits = [(A_inv*(x - B)) % 26 for x in digits]
                print("A = " + str(A) + " A^-1 = " + str(A_inv) + ", B = " + str(B) + ", A^-1*B = " + str(A_inv*B))
                
                #transform digits back to correct ASCII value
                digits = [(x + 65) for x in digits]
                
                #transform digits back to letters
                decrypted = [chr(ascii) for ascii in digits]
                
                #concatenate all the strings
                new_message = ""
                for i in range(len(decrypted)):
                    new_message = new_message + str(decrypted[i])
                
                print(new_message)
               
            
    return

affinedecrypt(message)
```
