```python
# Chapter 7, Exercise 7.6, polyalphabetic cryptosystem
# @Author = Skyler Moomey
# @Date = 10/21/2020
from array import*
import numpy as np

A = np.array([[3,4], [2,3]])
B = np.array([2,5])
     
# you can change this message, the program will encrypt whatever is within the quotations
message = "CRYPTOLOGY"
print("Unencrypted: " + message)

def poly_affine(message, A, B):
    
    #if message not even length, pad with X on end
    if len(message)%2!=0:
        message = message + "X"
    
    #convert message to into integer array
    int_message = np.array([ord(x) for x in message])

    #array of 2-tuples

    #instantiate with same num elements as int_message and |int_message|/2 rows, 2 columns
    p = np.arange(len(int_message)).reshape(len(int_message)/2, 2)

    #loop through elements of char_message
    #when x%2=0, put char_message[x] into p[x/2, 0], put char_message[x + 1] into p[x/2, 1]
    # example, when x=4, x%2=0. Then char_message[4] goes into p[2, 0] and char_message[5] goes into p[2, 1]
    for x in range(len(int_message)):
        if(x%2==0):
            p[x/2, 0] = int_message[x]
            p[x/2, 1] = int_message[x + 1]

    # subtract 65 from each 2-tuple
    six_five = np.array([65, 65])
    p = p - six_five

    #perform affine matrix transformation
    #i equals row-wise element in p
    #x equals single row vector in p
    i = 0
    for x in p[0:len(int_message)/2 - 1,]:
        x = np.matmul(A, x) + B
    
        #mod product
        x = x%26
        
        #assign results vector to p[i,] i.e. ith row of p
        p[i,] = x
        i+=1
    
    #add 65 to each 2-tuple
    p = p + six_five

    #reassign integer array with new elements
    #i equals index of int_message
    i=0
    for x in p[0:6, ]:
        #j equals index of element in row-wise element of p, only takes values {0,1}
        j=0
        int_message[i] = x[j]
    
        #increase int_message index
        i+=1
        int_message[i] = x[j+1]
    
        #increase int_message index
        i+=1
    #convert back to characters
    encrypted = [chr(x) for x in int_message]

    #put in string
    encrypted_str = ""
    for letter in encrypted:
        encrypted_str = encrypted_str + letter
        
    #return result
    return encrypted_str

#encrypt message
encrypted = poly_affine(message, A, B)
print("Encrypted: " + encrypted)

A_inv = np.array([[3,-4], [-2,3]])

A_invB = np.matmul(A_inv,B)*-1

print("Decryption formula is:\n f^-1(p) = \n" + str(A_inv) + "*p + " + str(A_invB))

#decrypt message
print("Decrypted: " + poly_affine(encrypted, A_inv, A_invB))
```
